# step2 collecting data sets

```python
song_df_1=pd.read_table("million_song.txt",header=None)
song_df_1.columns=['user_id','song_id','listen_count']
song_df_2=pd.read_csv("song_data.csv")
song_df=pd.merge(song_df_1, song_df_2.drop_duplicates(['song_id']),on="song_id",how="left")#merging both the data sets
song_df.head()
```
